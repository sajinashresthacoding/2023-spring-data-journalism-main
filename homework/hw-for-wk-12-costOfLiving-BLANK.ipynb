Each year the U.S. government increases the amount of social security retired people receive. That increase is based on the annual inflation rate  and the political climate (or party in control). I have indexed the numbers so 1995 is 100, and each year after that the number reflects a percentage increase since 1995. So between 1995 at 100 and 1996 at 102.6 reflects a 2.6 percent increase. And between 1995 and 1997 reflects a 5.6 percent increase.

```python
##note: so i have to do a roundabout and turn my nonetype into a dataframe
df = pd.DataFrame(df_raw)
df
```

```python
type(df)
```

```python
##finding pct chg for cable bill
old_cable = df["Avg. cable bill"].iat[0]
old_cable
```

```python
##finding pct chg for cable bill
df["cable_pct_chg"] = \
((df["Avg. cable bill"] - old_cable) / old_cable) * 100
df
```

```python
##finding pct chg for median household income

old_mhi = df["Median household income"].iat[0]
old_mhi
```

```python
df["mhi_pct_chg"] = \
((df["Median household income"] - old_mhi) / old_mhi) * 100
df
```

```python
## finding the accurate pct_chg for colA
index_colA = df["SS-colA"].iat[0]
index_colA
```

```python
df["cost_of_living"] = \
(df["SS-colA"] - index_colA)
df
```

```python
df.plot(x = "Year", y = ["cable_pct_chg", "mhi_pct_chg", "cost_of_living"],
       title = "Costs Over Time", style = {"cable_pct_chg": "red",
               "mhi_pct_chg": "blue", "cost_of_living": "gray"})
```
