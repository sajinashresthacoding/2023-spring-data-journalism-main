```python
import pandas as pd
```

```python
nc = pd.read_csv("311_Noise_Complaints.csv")
nc
```

```python
mi = pd.read_csv("median-income-2021-raw.csv")
mi
```

```python
##filter large median census data to smaller dataframe

midf = mi.filter(["NAME", "S1901_C01_012E"])
midf
```

```python
## rename columns
midf = midf.rename(columns={"NAME": "zip", "S1901_C01_012E": "median-income"})
midf
```

```python
midf = midf.drop(midf.index[0])
midf
```

```python
midf.info()
```

```python
midf["median-income"] = pd.to_numeric(midf["median-income"], errors = "coerce")
midf
```

```python
midf.info()
```

```python
midf["zip"] = midf["zip"].str.replace("ZCTA5 ", "")
midf
```

```python
midf.to_csv("median-income-2021.csv", index = False)
```
